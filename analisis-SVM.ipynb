```python
def clean_twitter_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # Remove user @ references and '#' from mentions
    text = re.sub(r'\@\w+|\#', '', text)
    
    # Remove punctuations
    text = re.sub(r'[^\w\s]', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    #menghapus retwet
    text = re.sub(r'RT[\s+]', '', text)
    
    return text
df['full_text'] = df['full_text'].apply(clean_twitter_text)
```

# Preprocessing

```python
# normalisasi
norm = {"yg" : "yang", "gmn" : "gimana", "cm" : "", "yuk" : "ayo", "tpi" : "tapi", "misii" : "permisi", "bgt" : "banget"}

def normalisasi(str_text):
    for i in norm:
        str_text = str_text.replace(i, norm[i])
    return str_text
df['full_text']=df['full_text'].apply(lambda x: normalisasi(x))
df
```

```python
# Stopword
import Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
more_stop_words = ['tidak']

stop_words = StopWordRemoverFactory().get_stop_words()
stop_words.extend(more_stop_words)

new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

def stopword(str_text):
    str_text = stop_words_remover_new.remove(str_text)
    return str_text

df['full_text'] = df['full_text'].apply(lambda x: stopword(x))
df.head()
```

```python
# Tokenisasi
df['tokenized'] = df['full_text'].apply(lambda x: x.split())
df.head()
```

```python
# Stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text_cleaning):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    do = []
    for w in text_cleaning:
        dt = stemmer.stem(w)
        do.append(dt)
    d_clean = []
    d_clean = " ".join(do)
    print(d_clean)
    return d_clean

df['stemmed'] = df['tokenized'].apply(stemming)
df.to_csv("hasilPreprocessing.csv", index=False)
data = pd.read_csv("hasilPreprocessing.csv", encoding='latin')
```

```python
#dataset = data.drop(['full_text'], axis=1, inplace=False)
#data = [tuple(x) for x in dataset.to_records(index=False)] 
# Mengonversi DataFrame data menjadi list of tuples
dataset = data.drop(['full_text', 'category', 'user_id_str', 'username', 'created_at', 'tokenized', 'stemmed', ], axis=1, inplace=False)
dataset = [tuple(x) for x in dataset.to_records(index=False)]
```

```python
import random
import math
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from textblob import TextBlob

# Filter out entries where tweet is NaN
dataset = [(tweet, label) for tweet, label in dataset if not (isinstance(tweet, float) and math.isnan(tweet))]

# Debug print to verify dataset after removing NaN values
print("Dataset after removing NaN values:")
for tweet, label in dataset:
    print(f"Tweet: {tweet}, Label: {label}")

# Create lists for positive, negative, and neutral tweets
set_positif = []
set_negatif = []
set_netral = []

# Separate tweets into positive, negative, and neutral lists
for n in dataset:
    if(n[1] == 'Positif'):
        set_positif.append((n[0], 'Positif'))  # Add tweet and label to positive list
    elif(n[1] == 'Negatif'):
        set_negatif.append((n[0], 'Negatif'))  # Add tweet and label to negative list
    else:
        set_netral.append((n[0], 'Netral'))  # Add tweet and label to neutral list

# Randomly sample half of the tweets from each list
set_positif = random.sample(set_positif, k=int(len(set_positif)/2))
set_negatif = random.sample(set_negatif, k=int(len(set_negatif)/2))
set_netral = random.sample(set_netral, k=int(len(set_netral)/2))

# Combine the sampled tweets into a single list
train = set_positif + set_negatif + set_netral

# Create a list of tuples for training
train_set = [(tweet, label) for tweet, label in train]

# Ensure test set is in the correct format
test_set = [(tweet, label) for tweet, label in dataset]

# Extract features and labels for training and testing
X_train, y_train = zip(*train_set)
X_test, y_test = zip(*test_set)

```

```python
import random
import math
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from textblob import TextBlob
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV, train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# TF-IDF Vectorizer
vectorizer = TfidfVectorizer(max_features=1000)

# Transform training and test data
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# SMOTE for handling class imbalance
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_tfidf, y_train)

# SVM classifier with GridSearchCV
svm_classifier = SVC()

pipeline = Pipeline([
    ('scaler', StandardScaler(with_mean=False)),
    ('classifier', svm_classifier)
])

param_grid = {
    'classifier__kernel': ['linear', 'rbf'],
    'classifier__C': [0.1, 1, 10],
    'classifier__gamma': ['scale', 'auto']
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')

# Train the classifier
grid_search.fit(X_train_res, y_train_res)

# Best parameters and best score
print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation accuracy: {:.2f}".format(grid_search.best_score_))

# Predict labels for train and test sets
train_predictions = grid_search.predict(X_train_res)
test_predictions = grid_search.predict(X_test_tfidf)

# Calculate accuracy for train and test sets
train_accuracy = accuracy_score(y_train_res, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

# Print the accuracy of the classifier
print("Akurasi Test: ", test_accuracy)


```

```python
# Evaluate other metrics
print("Evaluasi Metrik:")
print(classification_report(y_test, test_predictions))
```
